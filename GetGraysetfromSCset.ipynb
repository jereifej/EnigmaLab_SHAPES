```python
import numpy as np
import os
from matplotlib import pyplot as plt
import cv2

import shutil

# This notebook is used to transfer the samples from the sample list to "shape organized" groups
```

```python
def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
```

```python
dirTO = "ShapeGroups GRAY"
dirFROM = "Sample List Gray"
dirSEARCH = "ShapeGroups"
for root, dirs, files in os.walk(dirSEARCH):
    for file in files:
        getpath = find(file, dirFROM)
        subpath = root[12:18]
        pathTO = dirTO + '\\' + subpath
        path = path.replace('\\','/')
        pathTO = pathTO.replace('\\','/')
        print(pathTO)
        shutil.copy2(getpath, pathTO)
```
